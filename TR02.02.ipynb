```python
from Bio.PDB.MMCIF2Dict import MMCIF2Dict
mmcif_dict = MMCIF2Dict("3C5R.cif")
```

```python
import numpy as np
```

```python
label_seq_id = mmcif_dict['_atom_site.label_seq_id']
len(label_seq_id)
```

```python
auth_seq_id = mmcif_dict['_atom_site.auth_seq_id']
len(auth_seq_id)
```

```python
chain_id = mmcif_dict['_struct_asym.id']
print(chain_id)
```

```python
lookup = {}
for chain_h in chain_id:
    chain_lookup_h = {}
    for i in range(len(label_seq_id)):
        auth_seq_id_h = auth_seq_id[i]
        label_seq_id_h = label_seq_id[i]
        chain_lookup_h[auth_seq_id_h] = label_seq_id_h
    lookup[chain_h] = chain_lookup_h

len(lookup)
lookup["A"]["449"]
```
